```python
import pandas as pd

# 读取数据
try:
    df = pd.read_csv(r'c:\Users\32698\OneDrive\文档\课题会\data_R\processed_danshen_data.csv')
except FileNotFoundError:
    print("错误：找不到文件 'processed_danshen_data.csv'。请确认路径是否正确。")
    # Fallback for demonstration if file is not found
    df = pd.DataFrame() 

# 提取植物元素列
plant_element_cols = [col for col in df.columns if col.endswith('_P')]

if not plant_element_cols:
    print("错误：在文件中未找到以 '_P' 结尾的植物元素列。")
else:
    # 计算每种元素的平均含量
    mean_concentrations = df[plant_element_cols].mean()

    # 定义集合
    macro_elements = set()
    meso_elements = set()
    micro_elements = set()

    # 根据平均含量进行分类
    for col_name, mean_val in mean_concentrations.items():
        element_name = col_name.replace('_P', '') # 移除后缀
        if mean_val > 1000:
            macro_elements.add(element_name)
        elif 100 <= mean_val <= 1000:
            meso_elements.add(element_name)
        else: # < 100
            micro_elements.add(element_name)

    print("根据平均含量分类结果：")
    print("-" * 30)
    print(f"大量元素 (> 1000 mg/kg): {macro_elements or '无'}")
    print(f"中量元素 (100-1000 mg/kg): {meso_elements or '无'}")
    print(f"微量元素 (< 100 mg/kg): {micro_elements or '无'}")
```
